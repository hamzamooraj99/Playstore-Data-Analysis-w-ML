#Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#Import Data

```python
dataset = pd.read_csv('Final_App_Popularity_PK.csv')
```

#Genre

```python
genre_mean = dataset.groupby(['genre']).mean()['popularity_index']
```

```python
fig, ax = plt.subplots(figsize=(20, 18))

genre_popularity = pd.Series(genre_mean, index=genre_mean.index.values)
genre_popularity.sort_values(ascending=True, inplace=True)
genre_popularity.plot.barh(color='teal')
plt.rcParams['font.size']=18
plt.xlabel("Avg Popuarity Index")
plt.ylabel("Genre")
plt.title("Average Popularity per Genre - PK")
```

#Content Rating (Demographic)

```python
demographic_mean = dataset.groupby(['contentRating']).mean()['popularity_index']
```

```python
demographic_popularity = pd.Series(demographic_mean, index=demographic_mean.index.values)
demographic_popularity.sort_values(ascending=True, inplace=True)
demographic_popularity.plot.barh(color='teal')
plt.rcParams['font.size']=10
plt.xlabel("Avg Popuarity Index")
plt.ylabel("Content Rating")
plt.title("Average Popularity per Content Rating - PK")
```

#Offers IAP

```python
offers_iap_mean = dataset.groupby(['offersIAP']).mean()['popularity_index']
```

```python
iap_popularity = pd.Series(offers_iap_mean, index=offers_iap_mean.index.values)
iap_popularity.sort_values(ascending=True, inplace=True)
iap_popularity.plot.barh(color='teal')
plt.rcParams['font.size']=10
plt.xlabel("Avg Popuarity Index")
plt.ylabel("Offers IAP")
plt.title("Average Popularity per Offers IAP - PK")
```
